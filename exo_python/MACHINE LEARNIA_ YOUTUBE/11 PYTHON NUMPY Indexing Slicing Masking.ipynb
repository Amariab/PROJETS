```python
# technique de boolean indexing

A = np.random.randint(0, 10, [5, 5]) # des chiffres aléatoires qui vont de 0 à 10 dans un tableau de 5 sur 5
A
```

```python
# des opérations ==> élaboration d'un masque
A < 5
```

```python
# je choisis tous les numéros dont la valeur est en-dessous de 5 et ils vont être égaux à 10

A[A<5] = 10
A
```

```python
A = np.random.randint(0, 10, [5, 5]) 
A
```

```python
A [(A<5) & (A>2)] = 10
A
```
