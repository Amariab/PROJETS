```python
import matplotlib.pyplot as plt
import numpy as np # pour importer les fonctions mathématiques
from math import pi, cos

plt.clf() # efface la fenêtre graphique

# DEFINIR L'AXE DES ABSCISSES
x=np.linspace(0, 2*pi, 1000)
# numpy.linspace() permet d’obtenir un tableau 1D 
# allant d’une valeur de départ à une valeur de fin avec un nombre donné d’éléments.

y = [cos(t) for t in x] # tous les points de x auront pour image cos(t)

plt.plot(x, y, label = "cos(x)")
plt.title("Courbe du Cosinus")

plt.xlabel("abscisses")
plt.ylabel("ordonnées")

plt.legend()

plt.savefig("courbe_cos.png") # sauvegarde le graphique
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np # pour importer les fonctions mathématiques
from math import pi, cos

plt.clf() # efface la fenêtre graphique

# DEFINIR L'AXE DES ABSCISSES
x=np.linspace(0, 2*pi, 1000)
# numpy.linspace() permet d’obtenir un tableau 1D 
# allant d’une valeur de départ à une valeur de fin avec un nombre donné d’éléments.

y = [5*t+10 for t in x] # tous les points de x auront pour image cos(t)

plt.plot(x, y, label = "")
plt.title("Courbe")

plt.xlabel("abscisses")
plt.ylabel("ordonnées")

plt.legend()

plt.savefig("courbe1.png") # sauvegarde le graphique
plt.show()
```
